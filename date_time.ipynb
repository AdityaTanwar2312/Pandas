```python
import pandas as pd
from datetime import datetime
```

```python
#To convert to column to datetime format while loading the file use following method

df = pd.read_csv('data\ETH_1h.csv', parse_dates=['Date'], date_format='%Y-%m-%d %I-%p')
```

```python
df.head()
```

```python
df.loc[0, 'Date']
```

```python
#We need to provide a format in which our file contain date
#refer to the python date time documentation for each format as per the need 

# Link: http://bit.ly/pandas-dt-fmt

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %I-%p')
df['Date']
```

```python
#To get the day for a particular dat use day_name method

df.loc[0, 'Date'].day_name()
```

```python
#We can use dt class to implement function on the entire series

df['Date'].dt.day_name()
```

```python
df['DayOfWeek'] = df['Date'].dt.day_name()
df
```

```python
#Use min and max method to check the oldest and latest date

df['Date'].min()
```

```python
df['Date'].max()
```

```python
#To get time between two different dates use following method

df['Date'].max() - df['Date'].min()
```

```python
#To get data within a specific time range
# Note: Here we are using string to filter data in next example we will use datetime for filtering for required date

filt = (df['Date'] >= '2020')
df.loc[filt]
```

```python
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) & (df['Date'] < pd.to_datetime('2020-01-01'))
df.loc[filt]
```

```python
#We can set date as index value for ease of access

df.set_index('Date', inplace=True)
```
